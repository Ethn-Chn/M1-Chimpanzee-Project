```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene, ttest_ind, probplot, mannwhitneyu, wilcoxon
import seaborn as sns
```

```python
path = "data/DepE.csv"
DepE = pd.read_csv(path, delimiter=';')

# Convert the DataFrame values to float
for col in DepE.columns:
    try:
        DepE[col] = DepE[col].astype(float)
    except ValueError:
        pass
# Remove the last row of the DataFrame, it contains only NaN values
DepE = DepE[:20]
```

```python
# Calculate energy expenditures for the activity: Rest
SMER = DepE["Er"][:12] # Energy expenditures during no mais period
MER = DepE['Er'][12:20] # Energy expenditures during mais period

# QQ plot
fig, ax = plt.subplots()
x, y = probplot(SMER, plot=ax)
ax.get_lines()[0].set_markerfacecolor('#0E538F')
ax.get_lines()[1].set_color('#0E538F')
plt.title("Q-Q plot")
plt.xlabel("Quantiles théoriques")
plt.ylabel("Quantiles observés")
plt.show()

fig, ax = plt.subplots()
x, y = probplot(MER, plot=ax)
ax.get_lines()[0].set_markerfacecolor('#0E8F19')
ax.get_lines()[0].set_markeredgecolor('#0E8F19')
ax.get_lines()[1].set_color('#0E8F19')
plt.title("Q-Q plot")
plt.xlabel("Quantiles théoriques")
plt.ylabel("Quantiles observés")
plt.show()

# Shapiro-Wilk test to determine if the sample comes from a normally distributed population
statistic, p_value = shapiro(SMER)
print("Test de Shapiro : statistic :", statistic,"p-value  : ", p_value)
statistic, p_value = shapiro(MER)
print("Test de Shapiro : statistic :", statistic,"p-value  : ", p_value)

# Levene's test for equality of variances
statistic, p_value = levene(SMER, MER, center='median')
print("Test de Levene : statistic :", statistic,"p-value  : ", p_value)

# Student's t-test for comparing the means of two samples
statistic, p_value = ttest_ind(SMER, MER, equal_var=True)
print("Test de Student : statistic :", statistic, "p-value  : ", p_value)
```

```python
def load_pheno(filepath, month_name): # Load phenological data, extract necessary columns, and rename the 'rf_index' column.
    pheno = pd.read_excel(filepath)
    pheno_rf = pheno[['num_pheno_tree', 'rf_index']]
    pheno_rf = pheno_rf.rename(columns={'rf_index': f'rf_index_{month_name}'})
    return pheno_rf

# Reading FAI data and calculating basal area of trees
FAI = pd.read_excel("/content/drive/MyDrive/Stage M1/FAI/Disponibilité alimentaire_28032019.xlsx")
FAI['BA'] = 3.14 * (FAI['dbh_cm']/2)**2
df_FAI = FAI[['num_pheno_tree', 'BA']].copy()

# Loading phenological data for each month
months_data = {
    'Octobre': '/content/drive/MyDrive/Stage M1/Phénologie_EC/BDD_pheno_score_SBL_102018.xlsx',
    'Novembre': '/content/drive/MyDrive/Stage M1/Phénologie_EC/BDD_pheno_score_SBL_112018.xlsx',
    'Decembre': '/content/drive/MyDrive/Stage M1/Phénologie_EC/BDD_pheno_score_SBL_122018.xlsx',
    'Janvier': '/content/drive/MyDrive/Stage M1/Phénologie_EC/BDD_pheno_score_SBL_012019.xlsx'
}

for month, path in months_data.items():  # Merge phenological data with FAI data for each month.
    df_month = load_pheno(path, month)
    df_FAI = df_FAI.merge(df_month, on='num_pheno_tree', how='left')

# Add the species column
df_FAI["species"] = df_FAI['num_pheno_tree'].str[:3]

# Function to calculate FAI for each species
def calcul_FAI(group):
    denominator = group['BA'].sum() * 4
    for month in months_data.keys():
        numerator = (group['BA'] * group[f'rf_index_{month}']).sum()
        group[f'FAI_{month}'] = (numerator / denominator) * 100
    return group

# Apply the FAI calculation function to each species group
df_FAI = df_FAI.groupby('species', group_keys=False).apply(calcul_FAI)

# Aggregate FAI values for each species
FAI_per_species = df_FAI.groupby('species').agg({
    'FAI_Octobre': 'mean',
    'FAI_Novembre': 'mean',
    'FAI_Decembre': 'mean',
    'FAI_Janvier': 'mean'
}).reset_index()

FAI_per_species.head()
```

```python
# Comparing the Food Availability Index (FAI) between pairs of different months.
pairs = [("FAI_Octobre", "FAI_Novembre"),
         ("FAI_Octobre", "FAI_Decembre"),
         ("FAI_Octobre", "FAI_Janvier"),
         ("FAI_Novembre", "FAI_Decembre"),
         ("FAI_Novembre", "FAI_Janvier"),
         ("FAI_Decembre", "FAI_Janvier")]

wilcoxon_results = {}
# Conduct Wilcoxon tests for each pair of months
for pair in pairs:
    statistic, p_value = wilcoxon(FAI_per_species[pair[0]], FAI_per_species[pair[1]])
    wilcoxon_results[f"{pair[0]} vs {pair[1]}"] = p_value
# p-values for the comparisons of FAI between different pairs of months.
wilcoxon_results
```
